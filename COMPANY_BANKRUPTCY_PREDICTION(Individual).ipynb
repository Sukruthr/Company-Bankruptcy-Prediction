# **Problem Statement**

Prediction of bankruptcy is a phenomenon of increasing interest to firms who
stand to loose money because on unpaid debts. Since computers can store huge dataset
pertaining to bankruptcy making accurate predictions from them before hand is becoming
important. 

The data were collected from the Taiwan Economic Journal for the years 1999 to 2009. Company bankruptcy was defined based on the business regulations of the Taiwan Stock Exchange.

 In this project you will use various classification algorithms on bankruptcy
dataset to predict bankruptcies with satisfying accuracies long before the actual event.

### Handling Outliers 

```python
columns = df.columns
```

```python
# create box plot to show outliers

def boxplotFeatures(col):

  # exclude bankrupt col
  for i in col[1:]:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    df.boxplot(column = i, ax = ax)
    ax.set_ylabel(i)
    plt.show()

# call function
boxplotFeatures(columns)
```

```python
# data discrip before outlier removal
df.describe()
```

```python
# outlier capping using the IQR ranges
for col in columns[1:]:
  Q3 = df[col].quantile(0.75)
  Q1 = df[col].quantile(0.25)
  
  # calc IQR 
  IQR = Q3 - Q1
  df.loc[(df[col] > Q3 + (1.5*IQR)),col] = Q3 
  df.loc[(df[col] < Q1 - (1.5*IQR)),col] = Q1
```

### Model Building and Metric Calculations

```python
metric_report = pd.DataFrame(columns = ['logistic_regression','random_forest','SVM','gaussian_naive_bayes','KNN','XGBOOST','CATBOOST'], index = ['test_acc','train_acc','test_precision','train_precision','test_recall','train_recall','test_f1-score','train_f1-score'])
```

```python
# Generating Evaluation Matrix
def metricCalc(y_acc_test,y_acc_train,y_predicts_test,y_predicts_train,classifier_name):

  train_accuracy = accuracy_score(y_acc_train,y_predicts_train)
  test_accuracy = accuracy_score(y_acc_test,y_predicts_test)

  train_cm = confusion_matrix(y_acc_train,y_predicts_train)
  test_cm = confusion_matrix(y_acc_test,y_predicts_test )

  prec_train = train_cm[1][1]/(train_cm[1][1] + train_cm[0][1])
  prec_test = test_cm[1][1]/(test_cm[1][1] + test_cm[0][1])

  recall_train = train_cm[1][1]/(train_cm[1][0] + train_cm[1][1])
  recall_test = test_cm[1][1]/(test_cm[1][0] + test_cm[1][1])

  f1_train = (2*prec_train*recall_train)/(recall_train + prec_train)
  f1_test = (2*prec_test*recall_test)/(recall_test + prec_test)

  # auc_train = roc_auc_score(y_acc_train,y_train_prob)
  # auc_test = roc_auc_score(y_acc_test,y_test_prob)

  eval_list = [test_accuracy,train_accuracy,prec_test,prec_train,recall_test,recall_train,f1_test,f1_train]
  
  metric_report[classifier_name] = eval_list
```

```python
# function that runs various classification algos and calculates their respective eval metric
def model_result(train_X,test_X,train_y,test_y):

  # |||----------------------------------------------------------------------------------------------------------------------||| #

  # Logistic Regression:

  lr = LogisticRegression(max_iter = 10000)
  lr.fit(train_X,train_y)
  
  lr_class_train_pred = lr.predict(train_X)
  lr_class_test_pred = lr.predict(test_X)

  # lr_prob_train_pred = lr.predict_proba(train_X)[:,1]
  # lr_prob_test_pred = lr.predict_proba(test_X)[:,1]

# |||----------------------------------------------------------------------------------------------------------------------||| #

  # Descision Tree
  dt = DecisionTreeClassifier(criterion = 'entropy')
  dt.fit(train_X,train_y)
  # predict
  dt_class_train_pred = dt.predict(train_X)
  dt_class_test_pred = dt.predict(test_X)

  # dt_prob_train_pred = dt.predict_proba(train_X)[:,1]
  # dt_prob_test_pred = dt.predict_proba(test_X)[:,1]

# |||----------------------------------------------------------------------------------------------------------------------||| #

  # Random Forest

  rf = RandomForestClassifier(n_estimators = 60, criterion = 'entropy',random_state = 0)
  rf.fit(train_X,train_y)
  # predict
  rf_class_train_pred = rf.predict(train_X)
  rf_class_test_pred = rf.predict(test_X)

  # rf_prob_train_pred = rf.predict_proba(train_X)[:,1]
  # rf_prob_test_pred = rf.predict_proba(test_X)[:,1]

# |||----------------------------------------------------------------------------------------------------------------------||| #

  # SVM
  svc_clf = SVC(kernel = 'rbf')
  svc_clf.fit(train_X,train_y)
  # predict
  svc_class_train_pred = svc_clf.predict(train_X)
  svc_class_test_pred = svc_clf.predict(test_X)

  # svc_prob_train_pred = svc_clf.predict_proba(train_X)[:,1]
  # svc_prob_test_pred = svc_clf.predict_proba(test_X)[:,1]


# |||----------------------------------------------------------------------------------------------------------------------||| #

  # Naive Bayes
  nbc = GaussianNB().fit(train_X, train_y)
  # predict
  nbc_class_train_pred = nbc.predict(train_X)
  nbc_class_test_pred = nbc.predict(test_X)
  # nbc_prob_train_pred = svm_clf.predict_proba(train_X)[:,1]
  # nbc_prob_test_pred = svm_clf.predict_proba(test_X)[:,1]


# |||----------------------------------------------------------------------------------------------------------------------||| #

  # KNN
  knn = KNeighborsClassifier(n_neighbors = 3, metric = 'minkowski',p = 2)
  knn.fit(train_X, train_y)
  # predict
  knn_class_train_pred = knn.predict(train_X)
  knn_class_test_pred = knn.predict(test_X)

  # knn_prob_train_pred = knn.predict_proba(train_X)[:,1]
  # knn_prob_test_pred = knn.predict_proba(test_X)[:,1]


# |||----------------------------------------------------------------------------------------------------------------------||| #

# XGB
  xgb = XGBClassifier()
  xgb.fit(train_X, train_y)
  # predict
  xg_class_train_pred = xgb.predict(train_X)
  xg_class_test_pred =  xgb.predict(test_X)

  # xg_prob_train_pred =  xgb.predict_proba(train_X)[:,1]
  # xg_prob_test_pred =  xgb.predict_proba(test_X)[:,1]

# |||----------------------------------------------------------------------------------------------------------------------||| #

# CAT
  cat_ = CatBoostClassifier(verbose = 0)
  cat_.fit(train_X,train_y)
  # predict
  cat_class_train_pred = cat_.predict(train_X)
  cat_class_test_pred =  cat_.predict(test_X)

  # cat_prob_train_pred =  cat_.predict_proba(train_X)[:,1]
  # cat_prob_test_pred =  cat_.predict_proba(test_X)[:,1]


# Calculating the scores of all the models
  # metricCalc(test_y,train_y,lr_class_test_pred,lr_class_train_pred,'logistic_regression',lr_prob_test_pred,lr_prob_train_pred)
  # metricCalc(test_y,train_y,rf_class_test_pred,rf_class_train_pred,'random_forest',rf_prob_test_pred,rf_prob_train_pred)
  # metricCalc(test_y,train_y,svm_class_test_pred,svm_class_train_pred,'SVM',svm_prob_test_pred,svm_prob_train_pred)
  # metricCalc(test_y,train_y,nbc_class_test_pred,nbc_class_train_pred,'gaussian_naive_bayes',nbc_prob_test_pred,nbc_prob_train_pred)
  # metricCalc(test_y,train_y,knn_class_test_pred,knn_class_train_pred,'KNN',knn_prob_test_pred,knn_prob_train_pred)
  # metricCalc(test_y,train_y,xg_class_test_pred,xg_class_train_pred,'XGBOOST',xg_prob_test_pred,xg_prob_train_pred)
  # metricCalc(test_y,train_y,cat_class_test_pred,cat_class_train_pred,'CATBOOST',cat_prob_test_pred,cat_prob_train_pred)
  # 
  metricCalc(test_y,train_y,lr_class_test_pred,lr_class_train_pred,'logistic_regression')
  metricCalc(test_y,train_y,rf_class_test_pred,rf_class_train_pred,'random_forest')
  metricCalc(test_y,train_y,svc_class_test_pred,svc_class_train_pred,'SVM')
  metricCalc(test_y,train_y,nbc_class_test_pred,nbc_class_train_pred,'gaussian_naive_bayes')
  metricCalc(test_y,train_y,knn_class_test_pred,knn_class_train_pred,'KNN')
  metricCalc(test_y,train_y,xg_class_test_pred,xg_class_train_pred,'XGBOOST')
  metricCalc(test_y,train_y,cat_class_test_pred,cat_class_train_pred,'CATBOOST')


  # report
  return metric_report
```
